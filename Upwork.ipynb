```python
import numpy as np

# Define the noisy function
def f(x):
    return np.sin(x) + 0.1*np.random.randn(len(x))

# Define the smoothing window
window_size = 10

# Generate some x values
x = np.linspace(0, 2*np.pi, 100)

# Compute the noisy function values
y = f(x)

# Smooth the function by taking the moving average
y_smooth = np.convolve(y, np.ones(window_size)/window_size, mode='same')

# Compute the numerical derivative using the central difference method
dx = x[1] - x[0] # Assumes evenly spaced x values
derivative = (y_smooth[2:] - y_smooth[:-2]) / (2 * dx)

# Plot the results
import matplotlib.pyplot as plt
plt.plot(x, y, label='Noisy function')
plt.plot(x, y_smooth, label='Smooth function')
plt.plot(x[1:-1], derivative, label='Numerical derivative')
plt.legend()
plt.show()
```
